# Анализ факторов определяющих успешность игр
# Описание проекта
<p>В проекте проводится анализ факторов определяющих успешность игр на основе следующих данных: продажи, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Данные представляют собой выборку с 1980 по 2016 годы </p> 

**Цель**

- Выявление закономерностей, определяющих успешность игры, как основа для планирования кампании на 2017 год

**Задачи**

- Предобработка данных
- Выделение наиболее актуального периода
- Анализ платформ лидеров по продажам
- Оценка корреляции между объемом продаж и отзывами критиков и пользователей
- Определение предпочтений пользователей в зависти от региона (Япония, Европа, Северная Америка)
- Проверка рейтинговых гипотез


# Описание данных
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Предобработка данных

### Преобразование регистра

Переведем названия столбцов в нижний регистр, применив команду `str.lower()` к `data.columns`

```python
data.columns = data.columns.str.lower()
```

Проверим успешность преобразования

```python
data.columns
```

Переведем содержимое столбцы ___name, platform, genre и rating___ в нижний регистр, применив команду `str.lower()`. Преобразовани выполним внутри цикла

```python
columns = ['name', 'platform', 'genre', 'rating']
for i in columns:
    data[i] = data[i].str.lower()
```

Проверим успешность преобразования

```python
display(data.head())
```

Преобразование прошло успешно

Вновь проверим наличие дубликатов

```python
display(data.duplicated().sum())
```

### Преобразование типов

___year_of_release___ - год релиза, параметр не подразумевают дробной части, поэтому необходимо выполнить преобразования типов из строкового в целочисленный. Изменение типа реализуем методом `astype()`.
<br>Столбец содержит пропуски (NaN). Чтобы преобразования прошли успешно, заменим на тип `Int64`, а не просто `int`

```python
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

Проверим успешность преобразования

```python
data.dtypes
```

Преобразование прошло успешно

Параметр ___user_score___ - это оценка пользователей, но имеет строковый тип. Выведем на экран уникальные значения этого параметра

```python
display(data['user_score'].unique())
```

Если приглядеться, то можно увидеть значение 'tbd'. Аббревиатура 'tbd' означает, что данный параметр подлежит определению, т.е. оценка будет позже. Узнаем количество строк со значением 'tbd'

```python
display(len(data.query('user_score == "tbd"')))
```

Запись в формате 'tbd' будет только мешать нашему исследованию, лучше заменить ее на значение 'NaN'. Воспользуемся методом replace()

```python
data['user_score'] = data['user_score'].replace('tbd', np.nan)
```

Вновь выведем на экран уникальные значения параметра ___user score___ для проверки преобразований

```python
display(data['user_score'].unique())
```

Значения 'tbd' убрали, теперь преобразуем этот параметр в вещественный тип, так как оценка имеет дробную часть.

```python
data['user_score'] = data['user_score'].astype(float)
```

Проверим тип методом dtypes

```python
data.dtypes
```

Преобразование прошло успешно. Теперь типы всех столбцов соответствуют содержанию

Восстановить пропущенные значения можно на основе года выхода игр с тем же названием на других платформах. Однако, одна и таже игра могла выходить на разных платформах в разные годы. Пропуски в полях ___critic_score, user_score, rating___ также могут быть связаны с ошибками при заполнении таблицы или выгрузке данных. ### Как менялись продажи по платформам

Выведем на экран список платформ с общим объемом продаж. Список отсортируем по убыванию

```python
display(data
        .pivot_table(index='platform', values='total_sales', aggfunc='sum')
        .sort_values(by='total_sales',ascending=False)
        .reset_index()
        )
```

В списке 28 платформ. К самым прибыльным можно отнести первые шесть платфом, так как после 'ps' выручка резко снижается. Построим график с ежегодным выпуском игр для каждой платформы. Создадим переменную и присвоим ей сводную таблицу, в которой столбцы - платформы, а строки - годы релиза

```python
annual_release = data.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='count')
```

Отобразим первые пять строк табицы

```python
display(annual_release.head())
```

Используя инструменты библиотеки `matplotlib`, построим линейные графики по шести наиболее прибыльным платформам

```python
plt.figure(figsize=(9,6))
plt.plot(annual_release[['x360', 'wii', 'psp', 'ps3', 'ps2', 'ds']])
plt.title('Количество выпускаемых игр каждой платформой по годам')
plt.xlabel('Год релиза')
plt.ylabel('Количество игр')
plt.legend(['x360', 'wii', 'psp', 'ps3', 'ps2', 'ds'])
```

Оценим среднее значение продолжительности жизни компании

Построи 9 диаграмм рассеяния по одной на каждую платформу. Список платформ поместим в переменную `platform`. Автоматизируем процесс с помощью цикла. Подробное описание элементов цикла приведено в ячейке с кодом 

<a id='id4'></a>

```python
# Создадим набор "окон", в которых будем выполнять построение графиков
figure, axis = plt.subplots(3, 3, figsize=(20,30))

# Список названиями платформ
platform = data['platform'].unique()

# Создадим переменные, которые определяют позицию диаграммы
a = 0
b = 0

# Цикл перебирает элементы списка с названиями платформ
for pl_name in platform:
    
    # Условие, чтобы ограничить построения графиков в соответствии с матрицей окон 
    if a < 3 and b < 3:
        
        # Создадим диаграммы зависимости объема продаж от оценок критиков и пользователей
        axis[a,b].scatter(data=data[data['platform']== pl_name], x='user_score_norm', y='total_sales', alpha=0.7)
        axis[a,b].scatter(data=data[data['platform']== pl_name], x='critic_score', y='total_sales', alpha=0.7)
        
        # Зададим название, легенду, подписи осей и пределы оси x
        # Пределы оси y не задаем, иначе по некоторым платформам точки будут лежать на оси x
        axis[a,b].set_title(f'Платформа {pl_name}', fontsize= 19)
        axis[a,b].legend(['Пользователи', 'Критики'])
        axis[a,b].set_xlabel('Оценка критиков и пользователей', fontsize= 14)
        axis[a,b].set_ylabel('Суммарные продажи', fontsize= 14)
        axis[a,b].set_xlim(0,100)
        
        # Переход к следующему окну в данной строке
        b += 1
        
        # Условие перехода на новую строку окон
        if b == 3 and a < 6:
            a += 1
            b = 0

# Привественное сообщение-уведомление об окончании работы цикла    
print('Здравствуйте! Вы видете зависимость объема продаж от оценки критиков и пользователей.')
print('Все диаграммы рассеяния построены!')
```

Выше представлено 9 диаграмм рассеяния. На всех диаграммах для оси абсцисс установлены пределы от 0 до 100. Для оси ординат пределы не установлены, так как разброс по общему объему продаж у различных платформ внушительный. Среди графиков присутствует [диаграмма рассеяния для платформы ps3](#id1), она соответсвтует той, что мы строили ранее. Это подтверждает корректность построений. ***Коэффициент корреляции Пирсона для каждой платформы***

Рассчеты аналогичны тем, что мы делали ранее для [платформы ps3](#id2). Отличие в том, что результаты будут добавляться в соответствующие списки. Затем мы создадим датафрейм с результатами корреляции по платформам

```python
user_corr_series = []
critic_corr_series = []

for pl_name in platform:
    
    user_corr = round((data[data['platform']== pl_name]['total_sales']
                 .corr(data[data['platform']== pl_name]['user_score_norm'])
                 ),2
                )
    critic_corr = round((data[data['platform']== pl_name]['total_sales']
                   .corr(data[data['platform']== pl_name]['critic_score'])
                   ),2
                  )
    user_corr_series.append(user_corr)
    critic_corr_series.append(critic_corr)
```

Создадим датафрейм и отобразим его

```python
corr_all = pd.DataFrame({'platform':platform, 'user_corr':user_corr_series, 'critic_corr':critic_corr_series})
```

```python
plt.figure(figsize=(9,9))
plt.title('Корреляции объема продаж с оценками пользователей и критиков')
sns.lineplot(data=corr_all, x='platform', y='user_corr', label='Пользователи')
sns.lineplot(data=corr_all, x='platform', y='critic_corr', label='Критики')
plt.xlabel('Платформы')
plt.ylabel('Корреляция')
plt.ylim(-1,1)
plt.show()
```

Сделаем списки популярных в различных регионах жанров и платформ. Помимо списоков, построим диаграммы, отражающие распределение 5 наиболее популярных жанров и платформ в каждом регионе

```python
# Создадим необходимые списки
areas = ['na_sales', 'eu_sales','jp_sales']
genre_sales = []
platform_sales = []

# Создадим основу для построения графиков. Всего будет шесть окон, по три окна в двух строках
fig, axis = plt.subplots(2, 3, figsize=(18,18))

# Создадим переменные для указания положения диаграммы
a=0
b=0
c=1

# Внешний цикл переберает элементы списка areas, т.е. исследуемые регионы
for region in areas:
        
    # Выделяем 5 наиболее популярных жанров в регионе
    sorted_genre_by_sales = genre_in_areas.sort_values(by=region, ascending=False).head(5)

    # Выделяем 5 наиболее популярных платформ в регионе
    sorted_platform_by_sales = platform_in_areas.sort_values(by=region, ascending=False).head(5)
    
    # Добавим условия для преобразования названий регионов в более удачный формат
    if region == 'na_sales':
        region_name = 'Северная Америка'
    if region == 'eu_sales':
        region_name = 'Европа'
    if region == 'jp_sales':
        region_name = 'Япония'
    
    # Построим круговую диаграмму распсределения жанров по регионам
    axis[a,b].bar(sorted_genre_by_sales['genre'], sorted_genre_by_sales[region])
    axis[a,b].set_title(region_name)

    # Построим круговую диаграмму распсределения платформ по регионам
    axis[c,b].bar(sorted_platform_by_sales['platform'], sorted_platform_by_sales[region])
    axis[c,b].set_title(region_name)
    
    # Увеличим b на единицу, чтобы на следующей итерации построение выполнялось в следующем окне
    b += 1 
    
    # Внутренние циклы пробегают по значениям списков с топ 5 жанров/платформ
    # и добавляет по одному единые списки genre_sales и platform_sales
    for genre in sorted_genre_by_sales['genre']: 
        genre_sales.append(genre)      
       
    for platform in sorted_platform_by_sales['platform']:
        platform_sales.append(platform)
    
```

Мы создали два списка `genre_sales и platform_sales`, а также построили 6 круговых диаграмм. Диаграммы отражают распределение 5 наиболее популярных жанров и платформ в каждом регионе.  В каждом списке записано по 15 значений, первые пять значений соответствует топ 5 для Северной Америки, следующие пять - для Европы, а с 11 по 15 для Японии. Отобразим эти списки

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нулевая гипотеза формулируется так, чтобы использовать знак равенства. Другими словами, что все данные лежали в районе изначально предполагаемого значения

Сформулируем нулевую гипотезу:
>Средние пользовательские рейтинги платформ Xbox One и PC __одинаковые__

<br>Сформулируем альтернативную гипотезу:
>Средние пользовательские рейтинги платформ Xbox One и PC __разные__

Таким образом, у нас двусторонняя гипотеза о равенстве средних двух генеральных совокупностей. В этом случае для тестирования используют метод `ttest_ind` из библиотеки Scipy

Создадим переменные __xone_test и pc_test__ и присвоим им выборки с оценками пользователей на платоформах соответствующих платфомах. Удалим пропущенные значения методом `dropna()` и сбросим индексы `reset_index()`

```python
xone_test = data[data['platform']=='xone']['user_score'].dropna().reset_index(drop=True)
pc_test = data[data['platform']=='pc']['user_score'].dropna().reset_index(drop=True)
```

Рассчитаем и отобразим дисперсию по нашим выборкам

```python
display(np.var(xone_test))
display(np.var(pc_test))
```

Дисперсия составляет  1.1, что для шкалы диапазоном от 0 до 10 равняется примерно 10.1 %. Поэтому при тестировании гипотизы параметр `equal_var` установим равным - False. Переменная `alpha` - критический уровень статистической значимости. Присвоим ей значение 0.05, для нашей задачи этого достаточно. В задачах, где цена ошибки очень высока, например в медицине, этот параметр имеет смысл приравнивать 0.01

```python
alpha = .05 

results = st.ttest_ind(xone_test, pc_test, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

- Гипотеза о том, что средние пользовательские рейтинги платформ Xbox One и PC __одинаковые__ - `подтвердилась`

### Средние пользовательские рейтинги Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые. 

Сформулируем нулевую гипотезу:
>Средние пользовательские рейтинги жанров Action и Sports  __одинаковые__

<br>Сформулируем альтернативную гипотезу:
>Средние пользовательские рейтинги жанров Action и Sports __разные__

Таким образом, у нас двусторонняя гипотеза о равенстве средних двух генеральных совокупностей. В этом случае для тестирования используют метод `ttest_ind` из библиотеки Scipy

```python
action_test = data[data['genre']=='action']['user_score'].dropna().reset_index(drop=True)
sports_test = data[data['genre']=='sports']['user_score'].dropna().reset_index(drop=True)
```

Рассчитаем и отобразим дисперсию по нашим выборкам

```python
display(np.var(action_test))
display(np.var(sports_test))
```

Дисперсия большая, поэтому при тестировании гипотизы параметр `equal_var` сделаем равным - False. В этом случае при рассчетах будет использоваться критерий Уэлча.

```python
alpha = .05 

results = st.ttest_ind(action_test, sports_test, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

- Гипотеза о том, что средние пользовательские рейтинги жанров Action и Sports __одинаковые__ - `не подтвердилась`
